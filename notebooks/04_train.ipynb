```python
from ultralytics import YOLO
from pathlib import Path
```

```python
config_path = Path("../data/yolo/config.yaml")
model = YOLO('yolov8n.pt')
```

```python
results = model.train(
    data=str(config_path),
    epochs=20,
    imgsz=256,
    batch=16,
    name='coconut_tree_detection',
)
```
